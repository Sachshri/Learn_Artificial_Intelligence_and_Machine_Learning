```python
import pandas as pd
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
# Example data
data = {
    'gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male', 'Female'],
    'smoker': ['Yes', 'No', 'Yes', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)
# Create mosaic plot
mosaic(df, ['gender', 'smoker'], title='Mosaic Plot: Gender vs Smoker')

# Show plot
plt.show()
```
