```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Sample text data
text = """Data Visualization WordCloud( width=800, height=400, background_color="white"). generate(text)
"""

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color="pink").generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate synthetic exam scores data
np.random.seed(42)
exam_scores = np.random.normal(loc=58, scale=10, size=1000)
# Add some bimodality for illustration
# exam_scores = np.concatenate([exam_scores, np.random.normal(loc=85, scale=5, size=300)])

# Create a Kernel Density Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(exam_scores, shade=True, color="blue", bw_adjust=5)

# Add titles and labels
plt.title("Kernel Density Plot of Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Density")

# Show the plot
plt.show()
```
