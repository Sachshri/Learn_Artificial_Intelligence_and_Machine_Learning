## 1. Importing matplotlib and some necessary libraries <a id='Impoting'></a>

If we want to plot within an IPython notebook we can use `%matplotlib`. It comes with two different commands
1. %matplotlib notebook for interactive plots
2. %matplotlib inline for static plots

We will generally use `%matplotlib inline` 

```python
import numpy as np # type: ignore
import matplotlib as mp # type: ignore
import matplotlib.pyplot as plt # type: ignore

%matplotlib inline
```

#### 1.1 Generating Sin and Cos Graph using Matplotlib <a id='SinCosGraph'></a>

```python
x = np.linspace(0, 4*np.pi, 100)

# Defining Wave
sin_x = np.sin(x)
cos_x = np.cos(x)

# Plotting the Wave
plt.plot(x, sin_x, label = 'Sine')
plt.plot(x, cos_x, label = 'Cosine')

plt.legend(loc = 'best')
plt.title('Sine Wave')

plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 100)

# Defining Wave
sin_x = np.sin(x)
cos_x = np.cos(x)

# Plotting the Wave
plt.plot(x, sin_x,'*', label = 'Sine')
plt.plot(x, cos_x, '--',label = 'Cosine')

plt.legend(loc = 'best')
plt.title('Sine Wave and Cos wave')

plt.show()
```

1. `Legend` is used to add labels to the graph given in the plot function.  The loc argument takes an integer or string that specifies the location of the legend on the plot. different values for loc, such as <b>`lower right`, `center`, `upper right`, `lower left`</b>, etc.
2. `title` function adds a title to the plot

'-' will construct a line like in above plots

```python
plt.plot(x, sin_x,  '*')
plt.plot(x, cos_x,  '--')

plt.title('Sine and Cosine Wave')

plt.show()
```

<b>Note</b>
* In an IPython notebook the graph will be plotted even without using the plt.show() but in some conditions when we are working on big datasets and want to find some inference we use the plt.show() so it is generally a good practice to use it and also if we are writing a python script then also it's necessary to use the plt.show() command

## 4. Different Types of Plots <a id='DTOP'></a>

#### 4.1 Line Plot <a id='LP'></a>
A line plot is a type of plot used to visualize the relationship between two continuous variables. It is a plot that displays data points as connected line segments. Line plots are commonly used to visualize trends and patterns in data, especially when the data is continuous and has a large number of observations.

In a line plot, the x-axis represents one variable and the y-axis represents another variable. The data points are connected by lines to create a plot that shows how the values of the two variables change with respect to each other. Line plots are often used to visualize time-series data, where the x-axis represents time and the y-axis represents a measurement.

In Matplotlib, the <b>`plot`</b> function is used to create a line plot. You can specify the x and y values of the data points, along with other optional arguments such as <b> `color`, `line style`, and `markers` </b>. You can also add labels, titles, and legends to your line plot to provide context and clarity.


```python
x = np.linspace(0,4*np.pi, 100)
y = np.sin(x)

plt.figure(figsize=(16,5))
plt.style.use('Solarize_Light2')

plt.plot(x, y)

plt.title('Sine & Cosine Wave')

plt.show()
```

#### 4.1.2 Multiple Line Plots <a id='MLP'></a>

```python
x = np.linspace(0,4*np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize = (15,5))
plt.style.use('Solarize_Light2')

plt.plot(x, y_sin,  'o' ,alpha=0.6,color = 'red' , linewidth = 1, label = 'sin')
plt.plot(x, y_cos, '--' , color = 'blue', linewidth = 1, label = 'cos')
# plt.plot(x,y_sin)
plt.title('Sine & Cosine Wave')
plt.legend()
plt.show()
```

#### 4.1.4 We can also use linestyle command. It has 4 main styles solid, dashed, dashdot and dotted <a id='linestyle'></a>

```python
x = np.linspace(0, 20, 21, dtype = 'int')

plt.figure(figsize = (15,8))

plt.plot(x, x + 0, linestyle = 'solid'  , label = 'Solid')
plt.plot(x, x + 1, linestyle = 'dashed' , label = 'Dashed') 
plt.plot(x, x + 2, linestyle = 'dashdot', label = 'DashDot')
plt.plot(x, x + 3, linestyle = 'dotted' , label = 'Dotted')

plt.legend(loc = 'lower right')

plt.show()
```

#### 4.1.5 Adjusting the Plot Axes <a id='pltaxes'></a>
1. Can be done using <b>`plt.xlim`</b> for x axis 
2. <b>`plt.ylim`</b> for y axis

```python
x = np.linspace(0, 4*np.pi, 100)

plt.xlim(0,10)
plt.ylim(-2.5,0.6)

sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine Wave')

plt.show()
```

#### Can also be done using `plt.axis([xlim,ylim])`

```python
x = np.linspace(0, 4*np.pi, 100)

# plt.xlim([0,10])
# plt.ylim([-2.5,2.5])
plt.axis([0,10,-2.5,2.5])
sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine Wave')

plt.show()
```

#### To avoid adjust graph we can use `plt.axis('tight')` or `plt.axis('equal')` 

```python
x=np.linspace(0,4*np.pi,100)
plt.figure(figsize=(18,7))
sin_x=np.sin(x)
plt.style.use('Solarize_Light2')
plt.axis('equal')
plt.plot(x,sin_x,'o')
plt.show()
```

#### 4.1.6 Using `.set` to combine all of these commands <a id='set'></a>

```python
x = np.linspace(0, 4*np.pi, 100)

fig = plt.axes()

fig.plot(x, np.sin(x))

fig.set(xlim = (0,10), ylim = (-2,2), xlabel = 'x', ylabel = 'sin(x)', title = 'Sine Wave')

fig.plot()
```

### 4.2 Scatter Plots <a id='scatterPlot'></a>
A scatter plot is a type of data visualization in which individual data points are represented as points in a two-dimensional plane. The x-coordinate of each data point represents one feature of the data, and the y-coordinate represents another feature. Scatter plots are commonly used to visualize the relationship between two variables and to identify patterns or outliers in the data.

In Matplotlib, scatter plots can be created using the scatter function. The scatter function accepts two arrays of values, representing the x-coordinates and y-coordinates of the data points. You can also specify additional arguments to customize the appearance of the scatter plot, such as the size and color of the markers, the transparency of the markers, and the marker shape.

```python
n = 500
x = np.random.randn(n)
y = np.random.randn(n)

plt.scatter(x,y, c = 'red', alpha = .4, s = 30)

plt.title('Scatter Plot')

plt.show()
```

* In this code, the scatter function is used to create a scatter plot with 500 data points, represented as red markers with a size of 30 and a transparency of 0.5.

#### 4.2.2 Multiple Scatter Plots <a id='MSP'></a>

```python
n = 1000

x1 = np.random.randn(n)
y1 = np.random.randn(n)

x2 = np.random.randn(n) + 2
y2 = np.random.randn(n) + 2

plt.scatter(x1,y1, c = 'red',  alpha = .7, s = 10, label = 'Group - 1')

plt.scatter(x2,y2, c = 'blue', alpha = .7, s = 10, label = 'Group - 2')

plt.title('Scatter Plot with Multiple Groups')
plt.legend()
plt.show()
```

1. The primary difference between <b>`plt.scatter`</b> and <b>`plt.plot`</b> is that <b>`plt.scatter`</b> gives us the freedom to control the properties of each individual point (size,face_color,edge_color, transparency level etc)

```python
rng = np.random.RandomState(0)
plt.figure(figsize=(8,4))
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes  = rng.rand(100) * 1000

plt.scatter(x,y, c = colors, s = sizes , alpha = .7, cmap = 'viridis', edgecolors = 'black')

plt.title('Random Scatter Plot')
plt.xlabel('x values')
plt.ylabel('y values')

plt.colorbar()

plt.show()
```

* Alpha: Transparency of the markers in the scatter plot.
* Colorbar: A bar showing the colors used in the plot and the corresponding values.
* Colormap: A color mapping used to convert scalar data to colors.
* Edgecolors: The color of the edges of the markers in a scatter plot.
* Sizes: The size of the markers in a scatter plot.

## 4.3 Histograms and Binnings <a id='hist'></a>
 
A histogram is a plot that displays the distribution of a set of continuous or discrete data. The data is divided into a set of intervals or bins, and the height of each bar in the plot represents the number of data points that fall within that bin. Binnings refers to the process of dividing the data into these bins. The choice of bin size can greatly affect the appearance of the histogram, so it's important to choose a bin size that accurately represents the underlying distribution of the data.

```python
data = np.random.randn(1000)

plt.hist(data)

plt.show()
```

```python
data = np.random.randn(10000000)
plt.figure(figsize=(8,4))
plt.hist(data, bins = 2000, edgecolor = 'blue', alpha = .2, color = 'blue')

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

Individual Bins can be removed using <b>`histtype='stepfilled'`</b> 

```python
data = np.random.randn(10000000)
plt.figure(figsize=(8,4))
plt.hist(data, bins = 1000, edgecolor = 'Black', alpha = .8, color = 'grey', histtype = 'stepfilled')

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

#### 4.3.2 Multiple histograms <a id='Mph'></a>

```python
x1 = np.random.randn(1000)
x2 = np.random.randn(1000) + 2
x3 = np.random.randn(1000) + 4

plt.hist(x1, bins = 20,  alpha = .8)
plt.hist(x2, bins = 20,  alpha = .8)
plt.hist(x3, bins = 20, alpha = .8)

plt.title('Random Histogram')
plt.ylabel('Frequency')
plt.xlabel('Values')

plt.show()
```

## 4.4 2D Histograms
<b>`plt.hist2d`</b> is a function in matplotlib that generates a 2D histogram. A 2D histogram is a representation of the distribution of data over two variables. In contrast to a regular histogram which shows the distribution of a single variable, a 2D histogram shows the relationship between two variables. The function creates a heatmap that shows the density of data points in a 2D space.

```python
x = np.random.randn(1000)
y = np.random.randn(1000)
plt.figure(figsize=(7,4))
plt.hist2d(x,y, bins = 20 , cmap = 'viridis')

plt.title('2D Histogram')
plt.ylabel('x')
plt.xlabel('y')
plt.colorbar()
plt.show()
```

```python
plt.hist2d(x,y, bins = 20 , cmap = 'Blues')
plt.title('2D Histogram')
plt.ylabel('x')
plt.xlabel('y')
plt.colorbar()

plt.show()
```

## 4.5 Pie Charts <a id='pie'></a>
A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion.

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars,autopct='%.2f' )

plt.show()
```

#### 4.5.2 Basic operations <a id='Bo'></a>

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode)

plt.show()
```

`Adding Shadows`

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   10  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode, shadow = True)

plt.show()
```

`Adding Percentage`

```python
cars = ['AUDI','BMW','FORD','TESLA','JAGUAR','MERCEDES']
data = [ 450  , 385 , 35   ,   50  ,  200   , 500      ]
explode = [0, 0, 0 , 0, .1, 0] 

fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%')

plt.show()
```

`Adding Wedget`

```python
fig = plt.figure(figsize = (7,7))

plt.pie(data, labels = cars, explode = explode, shadow = True, autopct = '%2.2f%%',
        wedgeprops = {'edgecolor' : 'black', 'linewidth' : 3, 'antialiased' : True})

plt.show()
```

## 4.6 Multiple Subplots <a id='MSub'></a>
Multiple subplots refer to dividing a figure into multiple subplots or axes, each of which can contain a separate plot. This allows for multiple plots to be displayed in a single figure. In matplotlib, you can create multiple <b>`subplots`</b> using the subplot function.

`Index Based`

```python
x = np.linspace(0,10,50)
fig, ax = plt.subplots(2,2,figsize=(18,9))
y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')


y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')


y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')


plt.tight_layout()
plt.show()
```

`With Loops - Example 1`

```python
plt.subplots_adjust(hspace = .4, wspace = .4)

for i in range(1,7):
    
    plt.subplot(2,3,i)
    plt.text(0.5 , 0.5, str((2,3,i)), fontsize = 18, ha = 'center')
    
```

`With Loops - Example 2`

```python
x = np.linspace(0,10,50)
y = np.sin(x)


fig , (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title('Sin Wave')
ax2.set_title('Sin Wave - Square')
ax3.set_title('Sin Wave - Cube')

plt.tight_layout()

plt.show()
```
