# Data Visualization using matplotlib.pyplot

## Line Plot

```python
import numpy as np
import matplotlib.pyplot as plt 
x = np.linspace(0,100,10000)
y = np.cos(x)

plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Basic Line Plot')
plt.show()

```

```python
x=np.linspace(0,10,100)
y1=np.sin(x)
y2=np.cos(x)

plt.plot(x,y1,'o',linestyle='--',markersize=5,alpha=.6,label="sin(x)")
plt.plot(x,y2,'s',linestyle='-',color='pink',markersize=5,alpha=.6,label="cos(x)")
plt.legend(loc='lower left')
# plt.plot(x,y2)

plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [15, 18, 22, 28]

plt.plot(x, y1, label='Series 1')
plt.plot(x, y2, label='Series 2')
plt.title('Basic line plot')
plt.legend(loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Start', 'Sales', 'Returns', 'Marketing', 'R&D', 'Profit']
values = [1000, 300, -150, -200, -100, 850]  # Changes in values

# Calculate cumulative values
cumulative = np.cumsum(values)

# Create waterfall chart
plt.bar(categories, cumulative, color='gray')

# Highlight the positive and negative values differently
colors = ['green' if val >= 0 else 'red' for val in values]
plt.bar(categories, values, color=colors, edgecolor='black')

# Add labels
for i in range(len(categories)):
    plt.text(i, cumulative[i], f'{cumulative[i]:.0f}', ha='center',va='bottom')

# Add title and labels
plt.title('Waterfall Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Start', 'Sales', 'Returns', 'Marketing', 'R&D', 'Profit']
values = [1000, 300, -150, -200, -100, 850]  # Changes in values

# Calculate cumulative values
cumulative = np.cumsum(values)

# Shift cumulative values to align bars correctly
cumulative_shifted = np.hstack([[0], cumulative[:-1]])

# Create waterfall chart

# plt.bar(categories, values, bottom=cumulative_shifted, color='gray', edgecolor='black')
# Highlight positive and negative values
colors = ['green' if val >= 0 else 'red' for val in values]
plt.bar(categories, values, bottom=cumulative_shifted, color=colors, edgecolor='black')

# Add labels
for i in range(len(categories)):
    y_position = cumulative_shifted[i] + values[i]
    plt.text(i, y_position, f'{y_position:.0f}', ha='center', va='bottom' if values[i] >= 0 else 'top')

# Add title and labels
plt.title('Waterfall Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')

# Show plot
plt.show()
```

```python
#Spider plot radar plot
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['Sales', 'Marketing', 'R&D', 'Customer Support', 'Logistics']
values = [4, 3, 2, 5, 4]  # Scores for each category
values += values[:1]  # Close the loop by repeating the first value at the end

# Number of variables
num_vars = len(categories)

# Compute angle for each category
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Close the loop

# Plot the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

ax.fill(angles, values, color='blue', alpha=0.25)
ax.plot(angles, values, color='blue', linewidth=2)

# Add labels for each axis
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Set the range for each axis
ax.set_ylim(0, 5)

# Show plot
plt.title('Spider Chart Example')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for two groups
categories = ['Sales', 'Marketing', 'R&D', 'Customer Support', 'Logistics']
values_A = [4, 3, 2, 5, 4]
values_B = [3, 4, 5, 3, 2]

# Close the loop by repeating the first value at the end
values_A += values_A[:1]
values_B += values_B[:1]

num_vars = len(categories)

# Compute angle for each category
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Close the loop

# Plot the radar chart
fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

# Plot for Group A
ax.fill(angles, values_A, color='blue', alpha=0.25)
ax.plot(angles, values_A, color='blue', linewidth=2, label='Group A')

# Plot for Group B
ax.fill(angles, values_B, color='red', alpha=0.25)
ax.plot(angles, values_B, color='red', linewidth=2, label='Group B')

# Add labels for each axis
ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)

# Set the range for each axis
ax.set_ylim(0, 5)

# Add a legend
ax.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show plot
plt.title('Spider Chart with Multiple Datasets')
plt.show()
```

```python
# Sankey Diagram
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

# Create a simple Sankey diagram
sankey = Sankey(flows=[15, -5, -10, 10, -5, -5], 
                labels=['Input', 'Output 1', 'Output 2', 'Recycle', 'Waste', 'Loss'], 
                orientations=[0, 1, 1, -1, -1, 0])

# Finish and draw the diagram
sankey.finish()
plt.title('Simple Sankey Diagram')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey

# Create a customized Sankey diagram
sankey = Sankey(unit=None)

# Add first sub-diagram
sankey.add(flows=[15, -5, -10],
           labels=['Input', 'Output 1', 'Output 2'],
           orientations=[0, 1, -1],
           facecolor='green')

# Add a second sub-diagram connected to the first one
sankey.add(flows=[10, -5, -5],
           labels=['Recycle', 'Waste', 'Loss'],
           orientations=[-1, 1, 0],
           prior=0, connect=(2, 0), facecolor='orange')

# Finish and draw the diagram
sankey.finish()
plt.title('Customized Sankey Diagram with Sub-Flows')
plt.show()
```
